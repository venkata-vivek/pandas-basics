```python
# Find how many people knew python in one country usin filters

filt = df['Country'] == 'India'
df.loc[filt]['LanguageHaveWorkedWith'].str.contains('Python').sum()
```

```python
# Doing the same using groupby 
# We cannot do the below here , because this is not just a series this is a series group by object 
grouped_cuntry['LanguageHaveWorkedWith'].str.contains('Python').sum()
```

```python
# Using apply function o grou by series 
country_uses_python = grouped_cuntry['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum() )
```

```python
country_respondents =  df['Country'].value_counts()
country_respondents
```

```python
# Axis 1 = columns , axis 0 = Rows
# Concat methods combines 2 df's 
combined_df = pd.concat([country_respondents,country_uses_python],axis=1,sort=False)
combined_df
```

```python
combined_df.rename(columns={'count':'NumRespondents','LanguageHaveWorkedWith':'NumKnewPython'},inplace=True)
```

```python
# Formula to calculate Percentage
combined_df['PCTKnowsPython'] =  combined_df['NumKnewPython']/combined_df['NumRespondents'] * 100
```

```python
combined_df.sort_values(by='PCTKnowsPython',ascending=False,inplace=True)
combined_df
```
